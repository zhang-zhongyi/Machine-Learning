```python
#penultimate shots player distances
X['p.pen.base.corner'] = (((8.23/2)-X['player.distance.from.center'])**2 + 
                               (23.77/2-X['player.depth'])**2)**(1/2)
X['p.pen.basefar.corner'] = (((8.23/2)+X['player.distance.from.center'])**2 + 
                                  (23.77/2-X['player.depth'])**2)**(1/2)
X['p.pen.front.corner'] = (((8.23/2)-X['player.distance.from.center'])**2 + X['player.depth']**2)**(1/2)
X['p.pen.frontfar.corner'] = ((8.23/2+X['player.distance.from.center'])**2 + X['player.depth']**2)**(1/2)
X['p.pen.center'] = (X['player.distance.from.center']**2 + X['player.depth']**2)**(1/2)
X['p.pen.base']=(23.77/2)-X['player.depth']
X['p.pen.far.sideline']=(8.23/2)+X['player.distance.from.center']
X['p.pen.near.sideline']=(8.23/2)-X['player.distance.from.center']
```

```python
# penultimate shots opponent distances
X['o.distance.base.corner'] = (((8.23/2)-X['opponent.distance.from.center'])**2 + 
                               (23.77/2-X['opponent.depth'])**2)**(1/2)
X['o.distance.basefar.corner'] = (((8.23/2)+X['opponent.distance.from.center'])**2 + 
                                  (23.77/2-X['opponent.depth'])**2)**(1/2)
X['o.distance.front.corner'] = (((8.23/2)-X['opponent.distance.from.center'])**2 + X['opponent.depth']**2)**(1/2)
X['o.distance.frontfar.corner'] = ((8.23/2+X['opponent.distance.from.center'])**2 + X['opponent.depth']**2)**(1/2)
X['o.center'] = (X['opponent.distance.from.center']**2 + X['opponent.depth']**2)**(1/2)
X['o.base']=(23.77/2)-X['opponent.depth']
X['o.far.sideline']=(8.23/2)+X['opponent.distance.from.center']
X['o.near.sideline']=(8.23/2)-X['opponent.distance.from.center']
```

```python
#Distance between players
def pen_pen_distances(df):
    if df['same.side'] == 1:
        return ((df['opponent.distance.from.center']-df['player.distance.from.center'])**2+
               (df['opponent.depth']+df['player.depth'])**2)**(1/2)
    elif df['same.side'] == 0:
        return ((df['opponent.distance.from.center']+df['player.distance.from.center'])**2+
               (df['opponent.depth']+df['player.depth'])**2)**(1/2)

def end_pen_distances(df):
    if df['same.side'] == 1:
        return ((df['opponent.distance.from.center']-df['player.impact.distance.from.center'])**2+
               (df['opponent.depth']+df['player.impact.depth'])**2)**(1/2)
    elif df['same.side'] == 0:
        return ((df['opponent.distance.from.center']+df['player.impact.distance.from.center'])**2+
               (df['opponent.depth']+df['player.impact.depth'])**2)**(1/2)

X['pen.pen.distance'] = X.apply (lambda df: pen_pen_distances(df), axis=1)
X['end.pen.distance'] = X.apply (lambda df: end_pen_distances(df), axis=1)
```

```python
def ball_base_far_cornor(df):
    if (df['outside.sideline'] == 0) and (df['outside.baseline'] == 0 or 1):
        return ((8.23-df['distance.from.sideline'])**2+
               (df['depth'])**2)**(1/2)
    elif (df['outside.sideline'] == 1) and (df['outside.baseline'] == 0 or 1):
            return ((8.23+df['distance.from.sideline'])**2+
                   (df['depth'])**2)**(1/2)
X['end.ball.basefar.corner'] = X.apply(lambda df: ball_base_far_cornor(df), axis=1)
```

```python
def ball_front_cornor(df):
    if (df['outside.sideline'] == 0 or 1) and (df['outside.baseline'] == 0):
        return ((df['distance.from.sideline'])**2+
               (23.77/2-df['depth'])**2)**(1/2)
    elif (df['outside.sideline'] == 0 or 1) and (df['outside.baseline'] == 1):
            return ((df['distance.from.sideline'])**2+
                   (23.77/2+df['depth'])**2)**(1/2)
X['end.ball.front.corner'] = X.apply(lambda df: ball_front_cornor(df), axis=1)
```

```python
def ball_front_far_cornor(df):
    if (df['outside.sideline'] == 0 and df['outside.baseline'] == 0):
        return ((8.23-df['distance.from.sideline'])**2+
               (23.77/2-df['depth'])**2)**(1/2)
    elif (df['outside.sideline'] == 0 and df['outside.baseline'] == 1):
        return ((8.23-df['distance.from.sideline'])**2+
               (23.77/2+df['depth'])**2)**(1/2)
    elif (df['outside.sideline'] == 1 and df['outside.baseline'] == 0):
        return ((8.23+df['distance.from.sideline'])**2+
               (23.77/2-df['depth'])**2)**(1/2)
    elif (df['outside.sideline'] == 1 and df['outside.baseline'] == 1):
        return ((8.23+df['distance.from.sideline'])**2+
               (23.77/2+df['depth'])**2)**(1/2)
X['end.ball.frontfar.corner'] = X.apply(lambda df: ball_front_far_cornor(df), axis=1)
```

### Model 1 using RandomForestClassifier in sklearn.<a name="2a"></a>

[return to top](#home02)

```python
# import library
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV

# create a dictionary of parameters 
param_grid = {'n_estimators':list(range(700, 1100, 100)),
              'max_depth': [2,3],
              'max_features': [2,4,'sqrt'],
              'min_samples_split':[4,8], 
             }

# create Random Forest model
rf_obj = RandomForestClassifier(random_state=random_state)

# create gridsearch object with various combinations of parameters
model1_rnd_clf = GridSearchCV(rf_obj, param_grid, cv = 5, refit = True,
                                     n_jobs=-1, verbose = 5)

model1_rnd_clf.fit(X_train, y_train)
```

```python
# predict probabilities
y_pred_proba_model1_test = model1_rnd_clf.best_estimator_.predict_proba(X_test)

# predict classes
y_pred_model1_test = model1_rnd_clf.best_estimator_.predict(X_test)

print('First 5 predicted probabilities:\n', y_pred_proba_model1_test[0:5])
print('\nFirst 5 predictions:\n',y_pred_model1_test[0:5])
```

```python
# import library
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics

# mean error evaluation using the .score() method
model1_rdn_clf_score_test = model1_rnd_clf.best_estimator_.score(X_test, y_test)

model1_rdn_clf_confusion_matrix_test = metrics.confusion_matrix(y_test, y_pred_model1_test)

df_cm = pd.DataFrame(model1_rdn_clf_confusion_matrix_test, range(3), range(3))

sns.heatmap (df_cm, fmt='d',
             cmap='Blues_r', annot=True, square = True)

plt.xlabel('Predicted Class')
plt.ylabel('Actual Class')

model1_test_title = 'RandomForestClassifier - Confussion Matrix on Test Data\nMean Accuracy Score: {0:2f}'.format(model1_rdn_clf_score_test)
plt.title(model1_test_title, size = 14)
plt.show;
```

```python
# import library
from sklearn.metrics import classification_report

model1_rnd_clf_classification_report_test = classification_report(y_test, y_pred_model1_test)
print('\033[1m'+'  RandomForestClassifier - Classification Report on Test Data\n'+'\033[0m', model1_rnd_clf_classification_report_test)
```
